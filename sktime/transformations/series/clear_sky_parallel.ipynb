```python
from sktime.datasets import load_solar  # doctest: +SKIP
y = load_solar()
```

```python
y = y.to_frame()
y
```

```python
y['yday'] = y.index.day
y['hour'] = y.index.hour

# y_grouped_day = y.groupby(y.index.day)
# y_grouped_hour = y.groupby(y.index.hour)

y_grouped = y.groupby(['yday', 'hour'])

y_grouped
```
